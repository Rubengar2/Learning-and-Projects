## Dataset Description
This dataset contains performance statistics of football players across the FIFA WC tournament since 1966 to 2022. 

### Overview
- **Number of Players**: 
  - The dataset includes a total of 45 players.
  
- **Years Considered**:
  - The dataset covers matches from 1966 to 2022. 
    Please note that data before 1962 has been excluded due to the lack of clear statistics. Unfortunatelly we wont see good players as Luis Suarez, Garrincha, Didi or the first WC champions (30's and 50's)
  
- **Scope**:
  - Defensive and goalkeeper statistics have been omitted to focus on outfield player performance. 
    This decision was made to provide clarity in analysis and avoid biases. 
    However, an analysis focusing on defenders and goalkeepers can be conducted separately for a comprehensive understanding of team performance.


## Data Preprocessing
- The dataset has undergone preprocessing to remove goalkeeper-specific variables, as well as any incomplete or redundant data points. 
- Missing values and outliers have been addressed to ensure data quality and reliability.


```python
# List of goalkeeper-specific variables
goalkeeper_variables = ['saves', 'savedShotsFromInsideTheBox', 'savedShotsFromOutsideTheBox',
                        'goalsConcededInsideTheBox', 'goalsConcededOutsideTheBox', 'highClaims',
                        'punches']

# Drop the goalkeeper-specific variables from the DataFrame
df = df.drop(goalkeeper_variables, axis=1)

# Display the resulting DataFrame
df.head()
```

```python
# Create a new column 'Player_Year' by concatenating 'Player' and 'Year' columns to get better information in the future charts
df['Player_Year'] = df['player'] + ' - ' + df['year'].astype(str)

# Display the updated DataFrame
df.head()
```

```python
# Dictionary of color palettes for each World Cup-winning country from 1966 to 2022
paletas_paises_campeones = {
    'Brazil': ['#00843D', '#FFD700', '#00529C'],  # Green, Yellow, Blue
    'Germany': ['#000000', '#D3D3D3', '#0000FF', '#FF0000'],  # Black, Gray, Blue, Red
    'Italy': ['#00529C', '#FFFFFF', '#FF0000'],  # Green, White, Red
    'Argentina': ['#75AADB', '#F6D200', '#FFFFFF'],  # Sky Blue, Yellow, White
    'France': ['#0055A4', '#FFFFFF', '#EF4135'],  # Blue, White, Red
    'Uruguay': ['#75AADB', '#FFFFFF', '#FF0000'],  # Sky Blue, White, Red
    'England': ['#CC0000', '#FFFFFF', '#CC0000'],  # Blue, White, Red
    'Spain': ['#FF0000', '#FF0000', '#FFFF00'],  # Orange, Red, Yellow
    'West Germany' : ['#000000', '#D3D3D3', '#0000FF', '#FF0000'],  # Black, Gray, Blue, Red
}

# Save color palettes to a JSON file
import json

with open('paletas_paises_campeones.json', 'w') as file:
    json.dump(paletas_paises_campeones, file)

print("Color palettes saved successfully to 'paletas_paises_campeones.json'")
```

### **Fundamental Insights**

### Most goals in WC edition

```python
# Sort the DataFrame by the 'goals' column in descending order and take only the top 10 rows
df_top_10_goals = df.sort_values(by='goals', ascending=False).head(10)

# Create a color dictionary to map each team to a single color
team_colors = {team: color_palette[0] for team, color_palette in paletas_paises_campeones.items()}

# Create a bar plot using Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='goals', y='Player_Year', hue='team', data=df_top_10_goals, palette=team_colors, legend=False)

# Add the value inside each bar
for index, value in enumerate(df_top_10_goals['goals']):
    plt.text(value, index, str(value), va='center', ha='right', color='white', fontsize=11)

# Customize the plot
plt.xlabel('Goals')
plt.ylabel('Player - Year')
plt.title('Top 10 Goal Scorers per Edition of FIFA World Cup')

# Show the plot
plt.show()
```

This will lead me to analyze the results splited by metric in order to understand better which player was more dominant in every aspect considered during the analysis. 

```python
#Loading dataset
df_final = pd.read_csv('FWC_Stats3.csv')
df_final.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a color dictionary to map each team to a single color
team_colors = {team: color_palette[0] for team, color_palette in paletas_paises_campeones.items()}

# Get all numerical columns starting from the third column
columns_to_plot = df_final.iloc[:, 3:].select_dtypes(include=['float64', 'int64']).columns

# Loop over each column
for column in columns_to_plot:
    # Sort the DataFrame by the current column in descending order and take only the top 10 rows
    df_top_10 = df_final.sort_values(by=column, ascending=False).head(10)
    
    # Create a bar plot using Seaborn
    plt.figure(figsize=(12, 8))
    sns.barplot(x=column, y='player', hue='team', data=df_top_10, palette=team_colors, dodge=False)
    
    # Add the value inside each bar
    for index, value in enumerate(df_top_10[column]):
        plt.text(value, index, str(value), va='center', ha='right', color='white', fontsize=11)
    
    # Customize the plot
    plt.xlabel(column.capitalize())  # Capitalize the column name for the x-axis label
    plt.ylabel('Player - Year')
    plt.title(f'Top 10 Players by {column.capitalize()} per Edition of FIFA World Cup')
    plt.legend(loc='lower right')  # Move the legend to the lower right
    
    # Show the plot
    plt.show()
```
